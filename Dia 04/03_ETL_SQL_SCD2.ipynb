**CARGA DIMENSÃO CATEGORIA**

```sql
USE DW
GO
BEGIN TRANSACTION;  
	DECLARE @VDATA_INI DATETIME;
	SET @VDATA_INI= GETDATE();
BEGIN TRY  
--INICIA MERGE
INSERT INTO DW.SILVER2.DIM_CATEGORIA (Id_Categoria,NOME_CATEGORIA,DATA_INI,DATA_FIM,ATIVO) 
SELECT Id_Categoria,NOME_CATEGORIA,DATA_INI,DATA_FIM,ATIVO
FROM (
MERGE DW.SILVER2.DIM_CATEGORIA AS DESTINO

USING (SELECT * FROM DW.STAGING.STG_CATEGORIA_ERP) AS ORIGEM

ON DESTINO.[Id_Categoria] = ORIGEM.[Id_Categoria]
-- Há registro no destino e na origem mas com nomes diferente, update data fim
WHEN MATCHED 
AND (DESTINO.NOME_CATEGORIA <> ORIGEM.NOME_CATEGORIA)
AND DESTINO.ATIVO=1
THEN 
--ATUALIZA DATA FIM NO REGISTRO ANTIGO      
     UPDATE SET  DESTINO.DATA_FIM = GETDATE(),DESTINO.ATIVO=0

--Quando não há registro no destino e há na origem, insere
WHEN NOT MATCHED 
THEN 
    INSERT (Id_Categoria,NOME_CATEGORIA,DATA_INI,DATA_FIM,ATIVO) VALUES (ORIGEM.Id_Categoria,ORIGEM.NOME_CATEGORIA,GETDATE(),NULL,1)
-- Use os registros de origem que precisam ser inseridos no Dim por causa da atualização
 OUTPUT $action, 
    ORIGEM.Id_Categoria, 
    ORIGEM.NOME_CATEGORIA,
    getdate(),
    null,
	1
) 
as changes 
(
  action, 
  Id_Categoria,
  NOME_CATEGORIA,
  DATA_INI,
  DATA_FIM,
  ATIVO
)
where action='UPDATE';

INSERT INTO DW.LOGS.LOG_CARGAS2
           ([NUMERO_ERRO],
           [SERVERIDADE_ERRO],
           [ESTADO_ERRO],
           [PROC_ERRO],
           [LINHA_ERRO],
           [MSG_ERRO],
           [SITUACAO],
           [PROCESSO],
           [DATA_INI],
		   [DATA_FIM])
     VALUES
        (ERROR_NUMBER(),
        ERROR_SEVERITY(),
        ERROR_STATE(),
        ERROR_PROCEDURE(),
        ERROR_LINE() ,
        ERROR_MESSAGE() ,
        'SUCESSO',
       'CARGA CATEGORIA',
		@VDATA_INI,
		GETDATE()
		)

IF @@TRANCOUNT > 0  
    SELECT @@TRANCOUNT,'Sucesso';
    COMMIT TRANSACTION;  


--TERMINA MERGE
END TRY  
BEGIN CATCH
    SELECT @@TRANCOUNT,'ERRO';
 	IF @@TRANCOUNT > 0  
        ROLLBACK TRANSACTION; 

INSERT INTO[DW].[LOGS].[LOG_CARGAS2]
           ([NUMERO_ERRO],
           [SERVERIDADE_ERRO],
           [ESTADO_ERRO],
           [PROC_ERRO],
           [LINHA_ERRO],
           [MSG_ERRO],
           [SITUACAO],
           [PROCESSO],
           [DATA_INI],
		   [DATA_FIM])
     VALUES
        (ERROR_NUMBER(),
        ERROR_SEVERITY(),
        ERROR_STATE(),
        ERROR_PROCEDURE(),
        ERROR_LINE() ,
        ERROR_MESSAGE() ,
        'ERRO',
        'CARGA CATEGORIA',
		@VDATA_INI,
		GETDATE()
		)

        
END CATCH;  



 
```

```sql
USE DW
GO
IF NOT EXISTS
  (SELECT * FROM sysobjects WHERE name='FATO_VENDA_ERP' AND xtype='U')
CREATE TABLE GOLD2.FATO_VENDA_ERP
(
[SK_VENDA] INT IDENTITY(1,1) NOT NULL PRIMARY KEY,
[NUM_VENDA] INT,
[DATA_VENDA] DATE,
[SK_CLIENTE]INT,
[NUM_SEQ] INT,
[SK_PROD] INT,
[SK_VENDEDOR] INT,
[STATUS] CHAR(1),
[QTDE] DECIMAL(10,2),
[VAL_UNIT] DECIMAL(10,2),
[VAL_TOTAL] DECIMAL(10,2)
)
GO
```

**ANALISANDO AMOSTRAGEM DOS DADOS**

```sql
SELECT  
        A.NUM_VENDA,
        A.DATA_VENDA,
        E.SK_CLIENTE,
        B.NUM_SEQ,
        C.SK_PROD,
        D.SK_VENDEDOR,
        A.STATUS,
        B.QTDE,
        B.VAL_UNIT,
        B.VAL_TOTAL

  FROM DW.STAGING.STG_VENDAS_ERP A
  INNER JOIN DW.STAGING.STG_VENDA_ITENS_ERP B
  ON A.NUM_VENDA=B.NUM_VENDA
  
  INNER JOIN SILVER2.DIM_PRODUTOS C
  ON B.ID_PROD=C.ID_PROD
  AND C.ATIVO=1
   
  INNER JOIN SILVER2.DIM_VENDEDORES D
  ON A.ID_VENDEDOR=D.ID_VENDEDOR
  AND D.ATIVO=1

  INNER JOIN SILVER2.DIM_CLIENTES E
  ON A.ID_CLIENTE=E.ID_CLIENTE
  AND E.ATIVO=1
```

**CARGA FATO**

```sql
USE DW
GO
BEGIN TRANSACTION;  
   DECLARE @VDATA_INI DATETIME;
    DECLARE @VPROCESSO VARCHAR(30);
    SET @VDATA_INI= GETDATE();
    SET @VPROCESSO= 'CARGA FATO'
BEGIN TRY  

MERGE GOLD2.FATO_VENDA_ERP AS DESTINO

USING (
SELECT  
        A.NUM_VENDA,
        A.DATA_VENDA,
        E.SK_CLIENTE,
        B.NUM_SEQ,
        C.SK_PROD,
        D.SK_VENDEDOR,
        A.STATUS,
        B.QTDE,
        B.VAL_UNIT,
        B.VAL_TOTAL

  FROM DW.STAGING.STG_VENDAS_ERP A
  INNER JOIN DW.STAGING.STG_VENDA_ITENS_ERP B
  ON A.NUM_VENDA=B.NUM_VENDA
  
  INNER JOIN SILVER2.DIM_PRODUTOS C
  ON B.ID_PROD=C.ID_PROD
  AND C.ATIVO=1
   
  INNER JOIN SILVER2.DIM_VENDEDORES D
  ON A.ID_VENDEDOR=D.ID_VENDEDOR
  AND D.ATIVO=1

  INNER JOIN SILVER2.DIM_CLIENTES E
  ON A.ID_CLIENTE=E.ID_CLIENTE
  AND E.ATIVO=1 ) AS ORIGEM

ON (DESTINO.NUM_VENDA = ORIGEM.NUM_VENDA
AND DESTINO.NUM_SEQ = ORIGEM.NUM_SEQ)

-- Há registro no destino e na origem
WHEN MATCHED 

THEN 
    UPDATE SET    NUM_VENDA=ORIGEM.NUM_VENDA,
                  DATA_VENDA=ORIGEM.DATA_VENDA,
                  SK_CLIENTE=ORIGEM.SK_CLIENTE,
                  NUM_SEQ=ORIGEM.NUM_SEQ,
                  SK_PROD=ORIGEM.SK_PROD,
                  SK_VENDEDOR=ORIGEM.SK_VENDEDOR,
                  STATUS=ORIGEM.STATUS,
                  QTDE=ORIGEM.QTDE,
                  VAL_UNIT=ORIGEM.VAL_UNIT,
                  VAL_TOTAL=ORIGEM.VAL_TOTAL

--Quando não há registro no destino e há na origem
WHEN NOT MATCHED 

THEN 
   INSERT  
           (NUM_VENDA,DATA_VENDA,SK_CLIENTE,NUM_SEQ,SK_PROD,SK_VENDEDOR,STATUS,QTDE,VAL_UNIT,VAL_TOTAL)
     VALUES
           (      ORIGEM.NUM_VENDA,
                  ORIGEM.DATA_VENDA,
                  ORIGEM.SK_CLIENTE,
                  ORIGEM.NUM_SEQ,
                  ORIGEM.SK_PROD,
                  ORIGEM.SK_VENDEDOR,
                  ORIGEM.STATUS,
                  ORIGEM.QTDE,
                  ORIGEM.VAL_UNIT,
                  ORIGEM.VAL_TOTAL);

--FIM MERGE
-- GRAVA LOG
INSERT INTO [LOGS].[LOG_CARGAS2]
           ([NUMERO_ERRO],
           [SERVERIDADE_ERRO],
           [ESTADO_ERRO],
           [PROC_ERRO],
           [LINHA_ERRO],
           [MSG_ERRO],
           [SITUACAO],
           [PROCESSO],
           [DATA_INI],
		   [DATA_FIM])
     VALUES
        (ERROR_NUMBER(),
        ERROR_SEVERITY(),
        ERROR_STATE(),
        ERROR_PROCEDURE(),
        ERROR_LINE() ,
        ERROR_MESSAGE() ,
        'SUCESSO',
        @VPROCESSO,
		@VDATA_INI,
		GETDATE()
		)
SELECT 'SUCESSO ANTES',@@TRANCOUNT
IF @@TRANCOUNT > 0  
    COMMIT TRANSACTION;  
SELECT 'SUCESSO DEPOIS',@@TRANCOUNT    
END TRY  
BEGIN CATCH
SELECT 'ERRO ANTES',@@TRANCOUNT    
 	IF @@TRANCOUNT > 0  
        ROLLBACK TRANSACTION; 
SELECT 'ERRO DEPOIS',@@TRANCOUNT    
INSERT INTO [LOGS].[LOG_CARGAS2]
           ([NUMERO_ERRO],
           [SERVERIDADE_ERRO],
           [ESTADO_ERRO],
           [PROC_ERRO],
           [LINHA_ERRO],
           [MSG_ERRO],
           [SITUACAO],
           [PROCESSO],
           [DATA_INI],
		   [DATA_FIM])
     VALUES
        (ERROR_NUMBER(),
        ERROR_SEVERITY(),
        ERROR_STATE(),
        ERROR_PROCEDURE(),
        ERROR_LINE() ,
        ERROR_MESSAGE() ,
        'ERRO',
        @VPROCESSO,
		@VDATA_INI,
		GETDATE()
		)


END CATCH;  

```
